```python
#!pip install seaborn
#import seaborn as sb

#from __future__ import print_function
__author__= 'Oz'

import pylab as pl
#!pip install seaborn
#import seaborn as sb
#from scipy.optimize import curve_fit, minimize
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

%pylab inline

distributions = ['poisson', 'normal', 'chisq', 'cauchy', 'logistic', 'binomial']
distributions = ['chisq']
mymean = 100
print("mymean")
#numpy.random.chisquare(df, size=None)

df = mymean
md = {}
md['chisq'] = np.random.chisquare(df, size = 100)
pl.hist(md['chisq'], bins = 30)
pl.ylabel('N')
pl.xlabel('x')

```

```python
print("chisq mean: %.2f, standard deviation: %2f" %(md['chisq'].mean(), md['chisq'].std()))
```

```python
mysize = (2000 / (np.array(range(1, 100)))).astype(int)
print (mysize, mysize.shape)
```

```python
md['chisq'] = {} #Writin over discritionary entry created into a nested dict

for n in mysize:
    md['chisq'][n] = np.random.chisquare(df, size = n)
    
md['chisq']['means'] = {}
axchisq_mu_n = pl.figure(figsize=(10, 6)).add_subplot(111)

for nn in md['chisq']:
    if not type(nn) == str:
        md['chisq']['means'][nn] = md['chisq'][nn].mean()
        
        axchisq_mu_n.plot(nn, md['chisq']['means'][nn], 'o')
        axchisq_mu_n.set_xlabel('sample size', fontsize=18)
        axchisq_mu_n.set_ylabel('sample mean', fontsize=18)
        axchisq_mu_n.set_title('Figure-1: Chi Squared', fontsize=18)
        axchisq_mu_n.plot([min(mysize), max(mysize)], [df, df], 'k')
```

```python
# Describtion for Figure 2: Increasing in sample size leads to convegence in to the mean.
# law of large numbers states that as the number of trials increase, the actual
# results will converge over to the theoretical one. Mean 100 is the theoretical value here.
# and the plot here shows that about 600 samples the convergence with the mean 100
# takes place on the x-axis.
```

```python
allmeans = list(md['chisq']['means'].values())

pl.figure(figsize=(10, 10))
pl.hist(allmeans, bins=30)
pl.xlabel('sample mean', fontsize = 18)
pl.ylabel('N', fontsize = 18)
```
